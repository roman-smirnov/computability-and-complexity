```
# visualization imports 
import matplotlib.pyplot as plt
import networkx as nx

# initialize a directed graph
G=nx.DiGraph()

# init the nodes
G.add_nodes_from([0,1,2,3,4,5,6,7])
# init the edges
G.add_edges_from([(0,1),(0,2),(0,4),(0,5),(0,6),(0,7),(1,2),(2,3)])

# set node positioning - dot/graphviz is good f0r tree and state/trans plots
pos=nx.nx_pydot.graphviz_layout(G, prog='dot')

# init the labels
labels={0:r'$a$', 1:r'$b$',2:r'$c$',3:r'$d$',4:r'$\alpha$',
        5:r'$\beta$',6:r'$\gamma$',7:r'$\delta$'}
      
# set plot title
plt.title('Hierarchical Layout via Pydot')

# draw the graph on canvas
nx.draw(G, pos=pos,labels=labels, node_size=1200, font_size=32, arrowsize=20)

# remove plot axis
plt.axis('off')

# display the graph
plt.show()
```
